```python
df = df.drop(columns='Unnamed: 0', errors='ignore')
df
```

```python
df.info()
```

```python
#NOTE=replace function its is used to replace the value .and alsoused to remov the value
df['Ram']=df['Ram'].str.replace('GB','')#
df
```

```python
#NOTE it is used to convert the  datatype from one type to another type
df['Ram']=df['Ram'].astype('int64')
```

```python
df.info()
```

```python
df['Weight'] = df['Weight'].str.replace('kg','')
df['Weight']=df['Weight'].astype('float64')
df.info()
```
